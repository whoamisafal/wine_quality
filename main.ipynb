Basic Data Visualizations (Univariate Visualizations):

```python

red_wine_df.plot(kind='hist',title='Histogram of Red wine')
```

```python

white_wine_df.plot(kind='hist',title='Histogram of White wine')
```

```python
# extract columns
red_columns = list(red_wine_df.columns)
```

```python
plt.figure(figsize=(10,10))
for i in range(len(red_columns)):
    plt.subplot(4,3,i+1)
    sns.boxplot(data=red_wine_df[red_columns[i]])
    
```

Bivariate Visualizations:

```python
# Scatter Plot: 
plt.suptitle("Scatter plot of red wine ")
sns.scatterplot(red_wine_df,x='alcohol',y='quality')
```

```python
red_wine_df.columns
```

```python
# Scatter Plot: 
plt.suptitle("Scatter plot of white wine ")
sns.scatterplot(white_wine_df,x='pH',y='quality')
plt.show()
```

```python
# correlation 
correlation_red = red_wine_df.corr(numeric_only=True)
```

```python
correlation_red
```

```python
plt.suptitle("Correlation of the Red Wine")
sns.heatmap(data=correlation_red)
plt.show()
```

Red Wine 
The quality of wine is highly correlated with the alchol.
Correlation = 0.480343

```python
# correlation 
correlation_white = white_wine_df.corr(numeric_only=True)
correlation_white
```

```python
plt.suptitle("Correlation of the White wine")
sns.heatmap(data=correlation_white)
plt.show()
```
